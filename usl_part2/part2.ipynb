Goal is to build a synthetic data generation model using the existing data provided by the company.

Design a synthetic data generation model which can impute values [Attribute: Quality] wherever empty the company has missed recording the data.

```python
# importing relevant libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import KNNImputer
```

```python
# importing the dataset

df = pd.read_excel("company.xlsx")
```

```python
df.head()

# we can see nan in head
```

```python
df.describe()

# it appears that the feature set is in the same scale and therefore, scaling will not be required
```

```python
df.isnull().sum()
```

```python
df.info()

# there are 18 missing values in quality
```

```python
# exploring the values in quality

df.Quality.unique()
```

```python
# we will encode the values in quality to numbers so that the algorithm later can impute the missing values

df.Quality.replace({
    "Quality A": 1, # low quality
    "Quality B": 2 # high quality
}, inplace=True)
```

```python
# analysing the cluster behaviour before imputing

sns.pairplot(data=df,hue="Quality", palette="Set2",diag_kind="hist");

# the clusters are distinct and separable
```

```python
# we will use knn imputer to impute missing values in the target

imputer = KNNImputer()
df = pd.DataFrame(imputer.fit_transform(df),columns = df.columns)
```

```python
df.isnull().sum()
```

```python
df.info()

# there are no null values now
```

```python
sns.pairplot(data=df,hue="Quality", palette="Set2",diag_kind="hist");

# the separation and distinction is maintained after imputing missing values
```
