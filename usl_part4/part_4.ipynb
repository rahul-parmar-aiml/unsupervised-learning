```python
# defining a function to automate univariate numerical analysis

def univariateAnalysis_numeric(column,nbins):
    print("Description of " + column)
    print("----------------------------------------------------------------------------")
    print(df[column].describe(),end=' ')
    
    
    plt.figure()
    print("Distribution of " + column)
    print("----------------------------------------------------------------------------")
    sns.displot(df[column], kind='hist', color='g');
    plt.show()

    plt.figure()
    print("KDE of " + column)
    print("----------------------------------------------------------------------------")
    sns.kdeplot(df[column], color='g', fill=True);
    plt.show()
    
    plt.figure()
    print("BoxPlot of " + column)
    print("----------------------------------------------------------------------------")
    ax = sns.boxplot(x=df[column], color='g')
    plt.show()
    
    print("============================================================================")
```

```python
# creating a function to automate outlier treatment

def treat_outlier(x):
    # taking 5,25,75 percentile of column
    q5= np.percentile(x,5)
    q25=np.percentile(x,25)
    q75=np.percentile(x,75)
    dt=np.percentile(x,95)
    #calculationg IQR range
    IQR=q75-q25
    #Calculating minimum threshold
    lower_bound=q25-(1.5*IQR)
    upper_bound=q75+(1.5*IQR)
    #Capping outliers
    return x.apply(lambda y: dt if y > upper_bound else y).apply(lambda y: q5 if y < lower_bound else y)
```

2. Build a data driven model to rank all the players in the dataset using all or the most important performance features.

```python
# creting a copy of dataframe for use in KMeans
# we will cluster the dataset and group similar players together
# we will drop the name from the features set as it does not add value to the algorithm

X = df.drop("Name",axis=1).copy(deep=True)
```

```python
# importing required libraries

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
# scaling the data so that all the features are on the same scale, as Kmeans uses distance metric

scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
# we will us ethe elbow method, silhouette_score and try to find optimal number of clusters
# Checking for clusters ranging from 2 to 14 and view the impact on sum of squared errors

ssd = []
sil_coef = []

for k in range(2,15):
    
    model = KMeans(n_clusters=k, random_state=101)
    model.fit(scaled_X)
    ssd.append(model.inertia_)
    score = silhouette_score(scaled_X,model.labels_)
    sil_coef.append(score)
```

```python
# creating a graph to plot the sum of squared errors vs the number of clusters

plt.figure(figsize=(8,3),dpi=120)
plt.plot(range(2,15),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances");
```

```python
# creating a graph to plot the sum of squared errors vs the number of clusters

plt.figure(figsize=(8,3),dpi=120)
plt.plot(range(2,15),sil_coef,'o--')
plt.xlabel("K Value")
plt.ylabel("sil_coef");

# looking the two graphs for silhouette_score and ssd, we can see that there could be around 2 or 3 clusters, 
# but since the silhouette_score is dropping after two clusters, we will go ahead with 3 clusters
```

```python
# checking the loss in ssd during each cluster value

pd.Series(ssd).diff()
```

```python
# visualizing he above in bar plot
plt.figure(figsize=(8,3),dpi=120)
pd.Series(ssd).diff().plot(kind='bar');

# we can see that the drop from 3 to 4 has an slight elbow effect, indicating that this could be a good point to consider the cluster number
```

```python
# creating an Kmeans model with 3 clusters

model = KMeans(n_clusters=3, random_state=101)
model.fit(scaled_X)
```

```python
# adding the cluster labels back to the dataset X and we will check the feature corelations or the features that have highest impact on the clusters
X["K = 3"] = model.labels_
plt.figure(figsize=(8,3),dpi=120)
X.corr()['K = 3'].iloc[:-1].sort_values().plot(kind='bar');

# from the below graph, we can see tha the average has highest influence on the cluster formation
```

```python
# from the below table, we can see that the features

X.corr()["K = 3"].sort_values()[:-1]

# we see that the cluster are heavily inlfuenced by average
```

```python
# adding the name column back to the clustered dataset
X["name"] = df["Name"]
X.head(3)
```

```python
# groupng the dataset based on clusters

X.groupby('K = 3')[['Runs',"Ave","SR","Fours","Sixes","HF"]].mean()

# from the below table, we can see three groups of performance figures and we can categorize these into best, average and worst performers
```
